## Data from an Excel spreadsheet

Importing the data from an Excel file isn't much different that from a CSV file. Instead of `pd.read_csv()` we use `pd.read_excel()`.

```python
import plotly.express as px
import pandas as pd

url = 'https://raw.githubusercontent.com/Data-Dunkers/data-dunkers-modules/main/data-dunkers/Data/x-y-data.xlsx'
df = pd.read_excel(url)

# Create the plot
px.line(df, x='X', y='Y', title='Data from an Excel file')
```
