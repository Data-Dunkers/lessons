## Data from a table on a webpage

Instead of `read_csv()` or `read_excel()`, we are using `read_html()`, but one important difference is we have to specify which [table](https://www.w3schools.com/html/html_tables.asp) we want to use. 

When Pandas reads the tables on a webpage it indexes them starting from zero.

```python
import plotly.express as px
import pandas as pd

# Read the html file into a DataFrame named df
url = 'https://raw.githubusercontent.com/Data-Dunkers/data-dunkers-modules/main/data-dunkers/Data/x-y-data.html'
df = pd.read_html(url)[0]  # Index 0 is the first table

px.line(data_frame=df, x='X', y='Y', title='Data from a table on a webpage')
```

Note: if you have multiple rows for the column headers, look at **Fixing a multi-index** in this [cheat sheet](https://github.com/Data-Dunkers/lessons/blob/main/cheatsheet.md#fixing-a-multi-index).
