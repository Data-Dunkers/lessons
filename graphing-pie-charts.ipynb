# Creating Pie Charts

## Objectives

Students will be able to:

- Use the [Plotly Express](https://plotly.com/python/plotly-express/) library in [Python](https://www.python.org/) to create pie charts

## Importing and Filtering Data

We will use the [code library](https://en.wikipedia.org/wiki/Library_(computing)) called [pandas](https://pandas.pydata.org/) to import data from a [CSV file](https://en.wikipedia.org/wiki/Comma-separated_values).

```python
import pandas as pd
import plotly.express as px
url = 'https://raw.githubusercontent.com/Data-Dunkers/data-dunkers-modules/main/data-dunkers/Data/Pascal_Siakam.csv'
df = pd.read_csv(url)
df = df[df['SEASON_ID'] <= '2022-23']
df
```

## Pie Charts

When creating a [pie chart](https://plotly.com/python/pie-charts/), we use `values=` and `names=` instead of `x=` and `y=`.

```python
px.pie(df, values='FGM', names='SEASON_ID', title='Siakam Field Goals by Season')
```
